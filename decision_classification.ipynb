```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree

```

```python
balance_data = pd.read_csv('/DecisionTree_datatset (1).csv',
sep=',',header=0)
```

```python
balance_data.head()
```

```python
print("Dataset Length::"), len(balance_data)
print("Dataset Shape:: "), balance_data.shape
```

```python
print("Database::")
balance_data.head()
```

```python
x=balance_data.values[:,1:4]
y=balance_data.values[:,0]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
fitting_data = clf_entropy.fit(x_train,y_train)
```

```python
feature_names=balance_data.columns[1:4].tolist() #get feature names from column
class_names = balance_data.iloc[:,0].unique().astype(str).tolist() #get unique class names

plt.figure(figsize=(12,8))
tree.plot_tree(clf_entropy,feature_names=feature_names,class_names=class_names)
plt.show()
```

```python
y_pred_en = clf_entropy.predict(x_test)
y_pred_en
```

```python
predict = np.array([201,10018,250])
predict = predict.reshape(1,-1)
y_pred_en = clf_entropy.predict(predict)
y_pred_en
```

```python
print ("Accuracy is "), accuracy_score(y_test,y_pred_en)*100
```
